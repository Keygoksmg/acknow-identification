# The result of precision check. We manually filled up the columns of `result` and `isAffiliation`.
`result` are filled either by True, False, or NG. 
`isAffiliation` are filled either by True, False.
True in `isAffiliation` means the affiliation of acknowledged scholars is mentioned in acknowledgement text, while False is not.
True in `result` means that we assure the identified scholars with MAG dataset is correct, while False is not. NG means we cannot assure the correctness by poor information.

```python
import pandas as pd
```

```python
df = pd.read_csv('check_affiliation2.csv')

df['result'] = df.result.str.upper()

df
```

```python
stats = df['result'].value_counts()
stats
```

```python
stats['TRUE'] / (stats['TRUE'] + stats['FALSE'] + stats['NG']) * 100
```

```python
stats['TRUE'] / (stats['TRUE'] + stats['FALSE']) * 100
```
