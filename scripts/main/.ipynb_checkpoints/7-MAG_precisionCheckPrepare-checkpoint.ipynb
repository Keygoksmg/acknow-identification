# For precision check, we will randomly choise authors. 

```python
import ast
import pandas as pd
import numpy as np
import vaex
import os.path
from ast import literal_eval
from pathlib import Path

FILES = ['compbiology', 'biology', 'medicine', 'genetics', 'ntds', 'pathogenes', 'plosone', 'srep']

# Path to data
PATH_SAVED_CSV = '../../data-computed-MAG'
# PATH_MAG = '../../data-MAG/'
PATH_MAG = os.environ.get('DATA_MAG')

# Path to data to save
PATH_SAVE = '../../data-computed-MAG'
Path(PATH_SAVE).mkdir(parents=True, exist_ok=True) # create save dir if not exists
```

# Read MAG ID result

```python
# result of collaboration identifier
df_determinedAcknow_oneId = pd.read_csv(f'{PATH_SAVED_CSV}/d1Collaboration_result.csv')

# result of citation identifier
df_result_cited = pd.read_csv(f'{PATH_SAVED_CSV}/references_result.csv')
```

```python
# merge them
df_acknowId_d1collab_ref_merged = pd.merge(df_determinedAcknow_oneId, df_result_cited,  
          how='outer', on=['pid', 'acknow'])
# drop duplicated row
df_acknowId_d1collab_ref_merged = df_acknowId_d1collab_ref_merged.drop_duplicates(subset=['pid', 'acknow'])
df_acknowId_d1collab_ref_merged.head()
```

# create result dataset

```python
# Acknoweldged scholars identified only by collaboration approach
df_acknowId_from_d1Collab = df_acknowId_d1collab_ref_merged[df_acknowId_d1collab_ref_merged['ReferencedAuthorIds'].isnull()]
# Acknoweldged scholars identified only by citation approach
df_acknowId_from_ref = df_acknowId_d1collab_ref_merged[df_acknowId_d1collab_ref_merged['DetermindedAcknowId'].isnull()]
# Acknoweldged scholars identified by both collaboration and citation approach
df_acknowId_from_both = df_acknowId_d1collab_ref_merged[df_acknowId_d1collab_ref_merged['DetermindedAcknowId'] == df_acknowId_d1collab_ref_merged['ReferencedAuthorIds']]

# IDs of acknowledged scholars
df_acknowId = pd.concat([df_acknowId_from_d1Collab, df_acknowId_from_ref, df_acknowId_from_both])

# make new acknowId col
df_acknowId['acknowId'] = [ int(row['DetermindedAcknowId']) if np.isnan(row['ReferencedAuthorIds']) else (row['ReferencedAuthorIds']) for i, row in df_acknowId.iterrows()]
df_acknowId['acknowId'] = df_acknowId['acknowId'].astype(int)
df_acknowId.head()
```

#  Create Edgelist

### Merge with authors info of papers

```python
# read paperAuthorAffiliation of MAG
vaex_mag_pauthorAffil = read_vaex(f'{PATH_MAG}/PaperAuthorAffiliations/PaperAuthorAffiliations.csv')
```

```python
# summarise authors information of the acknowledged papers
vaex_row_paperinfo = vaex_mag_pauthorAffil[vaex_mag_pauthorAffil.PaperId.isin(df_acknowId.pid.tolist())]

# gropuby paper
df_row_paperinfo = vaex_row_paperinfo.to_pandas_df().rename({' AuthorId': 'AuthorId'}, axis=1)
df_row_paperinfo_groupby_paper = df_row_paperinfo.groupby('PaperId')['AuthorId'].apply(list).reset_index()
df_row_paperinfo_groupby_paper.head()
```

```python
df_acknow_wCollaborater = pd.merge(df_acknowId, df_row_paperinfo_groupby_paper,  
          how='inner', left_on='pid', right_on='PaperId')

df_acknow_wCollaborater.head()
```

### drop the cases the acknowledging author is the same as the acknoweldged author

```python
%%time
# drop the cases the acknowledging author is the same as the acknoweldged author
df_acknow_wCollaborater['acknowIsAuthor'] = df_acknow_wCollaborater.apply(lambda x: True if x['acknowId'] in x['AuthorId'] else False, axis=1) 
df_acknow_notAuthor = df_acknow_wCollaborater[~df_acknow_wCollaborater['acknowIsAuthor']]
df_acknow_notAuthor
```

# Check mismatching authors as far as possible

```python
# read Authors
vaex_mag_authors = vaex.open('../../data-MAG/Authors/Authors.csv.hdf5')
```

```python
# Author id list
author_ids = [i for l in df_acknow_notAuthor['AuthorId'] for i in l]

# extract possible acknow ids
vaex_author_ids = vaex_mag_authors[vaex_mag_authors['AuthorId'].isin(author_ids)]

# to dataframe
df_author_ids = vaex_author_ids.to_pandas_df().rename({' AuthorId': 'AuthorId'}, axis=1)[['AuthorId', 'DisplayName']]
df_author_ids.head()
```

```python
# explode `AuthorId` of df_acknow_notAuthor
df_acknow_notAuthor_exploded = df_acknow_notAuthor.explode('AuthorId')
```

```python
# explode `AuthorId` of df_acknow_notAuthor
df_acknow_notAuthor_exploded = df_acknow_notAuthor.explode('AuthorId')

# merge to display author name
df_acknow_notAuthor_exploded_wAuthorName = pd.merge(df_acknow_notAuthor_exploded, df_author_ids,  
          how='inner', left_on='AuthorId', right_on='AuthorId')

df_acknow_notAuthor_exploded_wAuthorName.head()
```

```python
len(df_acknow_notAuthor_exploded_wAuthorName)
```

### (last name is same as DisplayName?) & (initial of FirstName is the same?)

```python
# last name
df_acknow_notAuthor_exploded_wAuthorName['acknow_lastName'] = df_acknow_notAuthor_exploded_wAuthorName['acknow'].apply(lambda x: x.split(' ')[-1])
df_acknow_notAuthor_exploded_wAuthorName['author_lastName'] = df_acknow_notAuthor_exploded_wAuthorName['DisplayName'].apply(lambda x: x.split(' ')[-1])

# First Name initial
df_acknow_notAuthor_exploded_wAuthorName['acknow_firstInitial'] = df_acknow_notAuthor_exploded_wAuthorName['acknow'].apply(lambda x: x.split(' ')[0][0])
df_acknow_notAuthor_exploded_wAuthorName['author_firstInitial'] = df_acknow_notAuthor_exploded_wAuthorName['DisplayName'].apply(lambda x: x.split(' ')[0][0])
```

```python
# check
df_possible_duplication_AuthAndAck = df_acknow_notAuthor_exploded_wAuthorName[(df_acknow_notAuthor_exploded_wAuthorName['acknow_lastName'] == df_acknow_notAuthor_exploded_wAuthorName['author_lastName']) & 
                                        (df_acknow_notAuthor_exploded_wAuthorName['acknow_firstInitial'] == df_acknow_notAuthor_exploded_wAuthorName['author_firstInitial'])]
df_possible_duplication_AuthAndAck.head()
```

```python
len(df_possible_duplication_AuthAndAck)
```

### Papers_srep(plos)

```python
# read data
df_magpaper_id_doi_plos = pd.read_table(f'{PATH_MAG}/papers_plos.txt', sep=' ', names=['pid', 'doi'])
df_magpaper_id_doi_srep = pd.read_table(f'{PATH_MAG}/papers_srep.txt', sep=' ', names=['pid', 'doi'])
df_magpaper_id_doi = pd.concat([df_magpaper_id_doi_plos, df_magpaper_id_doi_srep])
# change doi to lower case
df_magpaper_id_doi["doi"] = df_magpaper_id_doi["doi"].str.lower()

# clear error row
s_vs = df_magpaper_id_doi['doi'].value_counts()
mag_available_dois = set(s_vs[s_vs.values == 1].index)
df_magpaper_id_doi2 = df_magpaper_id_doi[df_magpaper_id_doi['doi'].isin(mag_available_dois)]

df_magpaper_id_doi2.head()
```
